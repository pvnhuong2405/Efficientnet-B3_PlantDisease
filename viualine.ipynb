```python
import matplotlib.pyplot as plt
import cv2
import os
```

```python
url = "D:/Train_CNN/New Plant Diseases Dataset(Augmented)/New Plant Diseases Dataset(Augmented)/train"
```

```python
img_exts = (".jpg", ".jpeg", ".png", ".bmp", ".webp")

folder_counts = {}

for folder in os.listdir(url):
    folder_path = os.path.join(url, folder)
    if os.path.isdir(folder_path):
        count = sum(
            1 for f in os.listdir(folder_path)
            if f.lower().endswith(img_exts)
        )
        folder_counts[folder] = count

for k, v in folder_counts.items():
    print(f"{k}: {v} images")
```

```python
names = list(folder_counts.keys())
values = list(folder_counts.values())

plt.figure(figsize=(10, 5))
plt.bar(names, values)
plt.xticks(rotation=45, ha="right")
plt.ylabel("Number of images")
plt.title("Image count per folder")
plt.tight_layout()
plt.show()
```
